```python
import pandas as pd

df = pd.read_csv("athlete_events.csv")
df.head()
```

```python
import sqlite3

# Create SQLite in-memory database
conn = sqlite3.connect("olympics.db")

# Save DataFrame to SQL table
df.to_sql("athletes", conn, if_exists="replace", index=False)
```

```python
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())
```

```python
query = "SELECT * FROM athletes;"
pd.read_sql_query(query, conn)
```

How many olympics games have been held?

```python
query = """
SELECT COUNT(DISTINCT Year || '-' || Season) AS total_olympic_games
FROM athletes;
"""
pd.read_sql_query(query, conn)
```

List down all Olympics games held so far.

```python
query = """
SELECT DISTINCT Year, Season, City
FROM athletes
ORDER BY Year, Season;
"""
pd.read_sql_query(query, conn)
```

Mention the total no of nations who participated in each olympics game?

```python
query = """
SELECT Year, Season, COUNT(DISTINCT NOC) AS total_nations
FROM athletes
GROUP BY Year, Season
ORDER BY Year, Season;
"""
pd.read_sql_query(query, conn)
```

Which year saw the highest and lowest no of countries participating in olympics?

```python
query = """
WITH nation_counts AS (
    SELECT Year, Season, COUNT(DISTINCT NOC) AS total_nations
    FROM athletes
    GROUP BY Year, Season
)
SELECT *
FROM nation_counts
WHERE total_nations = (SELECT MAX(total_nations) FROM nation_counts)
   OR total_nations = (SELECT MIN(total_nations) FROM nation_counts);
"""
pd.read_sql_query(query, conn)
```

Which nation has participated in all of the olympic games?

```python
query = """
WITH all_games AS (
    SELECT COUNT(DISTINCT Year || '-' || Season) AS total_games
    FROM athletes
),
nation_participation AS (
    SELECT NOC, COUNT(DISTINCT Year || '-' || Season) AS games_participated
    FROM athletes
    GROUP BY NOC
)
SELECT NOC
FROM nation_participation, all_games
WHERE nation_participation.games_participated = all_games.total_games;
"""
pd.read_sql_query(query, conn)
```

Identify the sport which was played in all summer olympics

```python
query = """
WITH summer_games AS (
    SELECT COUNT(DISTINCT Year) AS total_summer_games
    FROM athletes
    WHERE Season = 'Summer'
),
sport_appearances AS (
    SELECT Sport, COUNT(DISTINCT Year) AS appearances
    FROM athletes
    WHERE Season = 'Summer'
    GROUP BY Sport
)
SELECT Sport
FROM sport_appearances, summer_games
WHERE sport_appearances.appearances = summer_games.total_summer_games;
"""
pd.read_sql_query(query, conn)
```

Which Sports were just played only once in the olympics?

```python
query = """
SELECT Sport
FROM (
    SELECT Sport, COUNT(DISTINCT Year || '-' || Season) AS appearances
    FROM athletes
    GROUP BY Sport
) AS sport_counts
WHERE appearances = 1;
"""
pd.read_sql_query(query, conn)
```

Fetch the total no of sports played in each olympic games.

```python
query = """
SELECT Year, Season, COUNT(DISTINCT Sport) AS total_sports
FROM athletes
GROUP BY Year, Season
ORDER BY Year, Season;
"""
pd.read_sql_query(query, conn)
```

Fetch details of the oldest athletes to win a gold medal.

```python
query = """
WITH gold_medalists AS (
    SELECT *
    FROM athletes
    WHERE Medal = 'Gold' AND Age IS NOT NULL
),
max_age AS (
    SELECT MAX(Age) AS oldest_age
    FROM gold_medalists
)
SELECT *
FROM gold_medalists
WHERE Age = (SELECT oldest_age FROM max_age);
"""
pd.read_sql_query(query, conn)
```

Find the Ratio of male and female athletes participated in all olympic games.

```python
query1 = """
SELECT
    Sex,
    COUNT(DISTINCT ID) AS athlete_count
FROM athletes
GROUP BY Sex
"""

pd.read_sql_query(query1, conn)
```

Fetch the top 5 athletes who have won the most gold medals.

```python
query2 = """
SELECT Name, Team, COUNT(*) AS gold_medals
FROM athletes
WHERE Medal = 'Gold'
GROUP BY Name, Team
ORDER BY gold_medals DESC
LIMIT 5
"""

pd.read_sql_query(query2, conn)
```

Fetch the top 5 athletes who have won the most medals (gold/silver/bronze).

```python
query3 = """
SELECT Name, Team, COUNT(*) AS total_medals
FROM athletes
WHERE Medal IN ('Gold', 'Silver', 'Bronze')
GROUP BY Name, Team
ORDER BY total_medals DESC
LIMIT 5
"""

pd.read_sql_query(query3, conn)
```

Fetch the top 5 most successful countries in olympics. Success is defined by no of medals won.

```python
query4 = """
SELECT Team, COUNT(*) AS total_medals
FROM athletes
WHERE Medal IN ('Gold', 'Silver', 'Bronze')
GROUP BY Team
ORDER BY total_medals DESC
LIMIT 5
"""

pd.read_sql_query(query4, conn)
```

List down total gold, silver and broze medals won by each country.

```python
query5 = """
SELECT
    Team,
    SUM(CASE WHEN Medal = 'Gold' THEN 1 ELSE 0 END) AS Gold,
    SUM(CASE WHEN Medal = 'Silver' THEN 1 ELSE 0 END) AS Silver,
    SUM(CASE WHEN Medal = 'Bronze' THEN 1 ELSE 0 END) AS Bronze
FROM athletes
GROUP BY Team
ORDER BY Gold DESC, Silver DESC, Bronze DESC
"""

pd.read_sql_query(query5, conn)
```

List down total gold, silver and broze medals won by each country corresponding to each olympic games.

```python
query5 = """
SELECT
    Team,
    SUM(CASE WHEN Medal = 'Gold' THEN 1 ELSE 0 END) AS Gold,
    SUM(CASE WHEN Medal = 'Silver' THEN 1 ELSE 0 END) AS Silver,
    SUM(CASE WHEN Medal = 'Bronze' THEN 1 ELSE 0 END) AS Bronze
FROM athletes
GROUP BY Team
ORDER BY Gold DESC, Silver DESC, Bronze DESC
"""

pd.read_sql_query(query5, conn)
```

Identify which country won the most gold, most silver and most bronze medals in each olympic games.

```python
query6 = """
SELECT
    Year,
    Team,
    SUM(CASE WHEN Medal = 'Gold' THEN 1 ELSE 0 END) AS Gold,
    SUM(CASE WHEN Medal = 'Silver' THEN 1 ELSE 0 END) AS Silver,
    SUM(CASE WHEN Medal = 'Bronze' THEN 1 ELSE 0 END) AS Bronze
FROM athletes
GROUP BY Year, Team
ORDER BY Year, Gold DESC
"""

pd.read_sql_query(query6, conn)
```

Identify which country won the most gold, most silver, most bronze medals and the most medals in each olympic games.

```python
# For Gold
query7_gold = """
WITH GoldMedals AS (
    SELECT Year, Team, COUNT(*) AS Gold
    FROM athletes
    WHERE Medal = 'Gold'
    GROUP BY Year, Team
)
SELECT *
FROM GoldMedals gm
WHERE Gold = (
    SELECT MAX(Gold)
    FROM GoldMedals
    WHERE Year = gm.Year
)
ORDER BY Year
"""

pd.read_sql_query(query7_gold, conn)
```

Which countries have never won gold medal but have won silver/bronze medals?

```python
query8 = """
SELECT Team
FROM athletes
WHERE Medal IN ('Silver', 'Bronze')
AND Team NOT IN (
    SELECT DISTINCT Team
    FROM athletes
    WHERE Medal = 'Gold'
)
GROUP BY Team
"""

pd.read_sql_query(query8, conn)
```

In which Sport/event, India has won highest medals.

```python
query9 = """
SELECT Sport, COUNT(*) AS total_medals
FROM athletes
WHERE Team = 'India' AND Medal IN ('Gold', 'Silver', 'Bronze')
GROUP BY Sport
ORDER BY total_medals DESC
LIMIT 1
"""

pd.read_sql_query(query9, conn)
```

Break down all olympic games where india won medal for Hockey and how many medals in each olympic games.
